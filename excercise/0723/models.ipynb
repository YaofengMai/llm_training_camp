生成英文文本

```python
data = openai.Completion.create(
  model="text-davinci-003",
  prompt="Say this is a greeting",
  max_tokens=7,
  temperature=0
)
print(data)
```

```python
print(data['choices'][0]['text'].strip())
```

聊天机器人

```python
messages=[
    {
        "role": "user",
        "content": "Hello!"
    }
]


data = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages = messages
)
print(data)
```

```python
new_message = data['choices'][0]['message']
messages.append(new_message.to_dict())
print(messages)
```

```python
new_chat = {
    "role": "user",
    "content": "1.讲一个程序员才听得懂的冷笑话；2.今天是几号？3.明天星期几？"
}


messages.append(new_chat)
```

```python
data = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=messages
)
```

```python
new_message = data['choices'][0]['message']
messages.append(new_message.to_dict())
print(messages)
```

多角色对话

```python
import openai

data = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "你是一个助理"},
        {"role": "user", "content": "谁赢了2018年足球世界杯冠军"},
        {"role": "assistant", "content": "法国赢了2018年足球世界杯冠军"},
        {"role": "user", "content": "该比赛在哪里举办?"}
    ]
)
```

```python
message = data["choices"][0]["message"]["content"]
print(message)

```
