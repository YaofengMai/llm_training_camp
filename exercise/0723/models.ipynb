```python
messages=[
    {
        "role": "user",
        "content": "Hello!"
    }
]


data = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages = messages
)
print(data)
```

```python
new_message = data['choices'][0]['message']
messages.append(new_message.to_dict())
print(messages)
```
